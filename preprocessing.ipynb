```python
from datetime import datetime
from tqdm import tqdm
type_zone = crimes_no_na_df[['Community Area', 'Primary Type', 'Date']]
#d = type_zone['Date'].values
#day = []
#month = []
#year = []
date1 = []
for i in type_zone['Date'].values:
    v = i.split(' ', -1)
    date1.append(v[0])
#    vals = v[0].split('/', -1)
#    day.append(int(vals[1]))
#    month.append(int(vals[0]))
#    year.append(int(vals[2]))
type_carea_df = pd.DataFrame({'Date': date1, 'Community Area': type_zone['Community Area'].values, 'Primary Type': type_zone['Primary Type'].values})
test = type_carea_df.groupby(type_carea_df.columns.tolist(),as_index=False).size()
#print(test.size)
dates = type_carea_df['Date'].unique()
#careas = type_carea_df['Community Area'].unique()
#types = type_carea_df['Primary Type'].unique()
df_date = []
df_carea = []
df_type = []
df_num = []
for d in tqdm(sorted(dates, key=lambda date: datetime.strptime(date, "%m/%d/%Y"))):
    for c in test[d].keys():
        c_id = c[0]
        if (isinstance(test[d][c_id], np.int64)):
            continue
        for k in test[d][c_id].keys():
            df_date.append(d)
            df_carea.append(c_id)
            df_type.append(k)
            df_num.append(test[d][c_id][k])
print(len(df_date))
print(len(df_carea))
print(len(df_type))
print(len(df_num))
final_df = pd.DataFrame({'Date': df_date, 'Community Area': df_carea, 'Primary Type': df_type, 'Number': df_num})
final_df.to_csv(os.path.join(dataset_path, 'crimes_by_type_by_carea.csv'))
print("done")
#for i in test:
#    print("i : %d"%i)
#    print(test[i])
#print(test[77]['ASSAULT'])
#print(careas)
#print(len(careas))
#for n in range(len(careas)):
    
#print(type_carea_df.duplicated())
#print(types)
#type_carea_df.to_csv(os.path.join(dataset_path, 'crimes_by_type_by_carea.csv'))
```
